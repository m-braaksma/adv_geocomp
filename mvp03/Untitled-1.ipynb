```python
import osmnx as ox
import pandas as pd

def get_smallest_admin_level(country_name):
    # Step 1: Get the country boundary polygon
    country_boundary = ox.geocode_to_gdf(country_name)
    
    # Step 2: Query for all administrative boundaries within the country
    tags = {'boundary': 'administrative'}
    boundaries = ox.geometries.geometries_from_polygon(
        country_boundary.unary_union, tags
    )
    
    # Step 3: Extract the admin_level information
    boundaries['admin_level'] = pd.to_numeric(boundaries['admin_level'], errors='coerce')
    
    # Step 4: Find the smallest admin_level
    smallest_level = boundaries['admin_level'].min()
    
    # Step 5: Filter and display boundaries with the smallest admin_level
    smallest_boundaries = boundaries[boundaries['admin_level'] == smallest_level]
    
    return smallest_level, smallest_boundaries

# Example usage for Kenya
country = "Belize"
smallest_level, smallest_boundaries = get_smallest_admin_level(country)
print(f"The smallest admin_level in {country} is: {smallest_level}")

# Display the resulting boundaries
print(smallest_boundaries[['name', 'admin_level']].head())
```

```python
import overpy

api = overpy.Overpass()

# Query to fetch all administrative boundaries in Belize
query = """
[out:json];
area["name"="Belize"]["boundary"="administrative"]->.searchArea;
relation["boundary"="administrative"](area.searchArea);
out tags;
"""
result = api.query(query)

# Print out admin levels found
admin_levels = set()
for relation in result.relations:
    admin_level = relation.tags.get("admin_level")
    if admin_level:
        admin_levels.add(admin_level)
        print(f"Name: {relation.tags.get('name')}, Admin Level: {admin_level}")

print(f"Admin levels found: {sorted(admin_levels)}")
```
